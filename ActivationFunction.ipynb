```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def s(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.linspace(-10, 10, 50)
X
```

```python
plt.plot(X, s(X), 'b')
plt.grid()
plt.text(2, 0.37, r'$\sigma(x)=\frac{1}{1+e^{-x}})$', fontsize=20, color='green')
```

# Relu

```python
def relu(x):
    return np.maximum(0.0, x)
```

```python
X = np.linspace(-10,10,50)
plt.plot(X, relu(X), 'r')
plt.grid()
plt.text(-9, 6, r'$EwlU(X)=max(0.0, x)', fontsize=20)
```
