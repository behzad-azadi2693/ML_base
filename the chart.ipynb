### histogram

```python
data = np.array(list(np.random.randint(0,20, 40)))
```

```python
plt.hist(data, bins='auto',edgecolor='black');
```

```python
data = 100 +15 * np.random.randn(1000)
n,bins,_ = plt.hist(data, bins=20, edgecolor='yellow')
y = norm.pdf(bins, statistics.mean(data), statistics.stdev(data))
plt.plot(bins,y,c='r');
```
